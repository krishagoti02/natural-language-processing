```python
import requests
link = 'https://writingcenter.unc.edu/tips-and-tools/paragraphs/'
page = requests.get(link)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
table = soup.find_all('p')
text = [i.text for i in table]
print(text)
```

```python
from pprint import pprint
pprint(text)
```
